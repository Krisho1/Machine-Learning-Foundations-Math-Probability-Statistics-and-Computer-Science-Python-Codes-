##Calculus 1 (Limit and Derivatives)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,1000)
```

```python
y = x**2 + 2*x + 2
```

```python
fig,ax = plt.subplots()
_ = ax.plot(x,y)
```

```python
fig,ax = plt.subplots()
xlim = (-2,0)
ylim = (0,2)
_=ax.set_xlim(xlim)
_=ax.set_ylim(ylim)
_=ax.plot(x,y)
```

```python
fig,ax = plt.subplots()
_=ax.set_xlim([-1.1,-0.9])
_=ax.set_ylim([0.9,1.1])
ax.plot(x,y)
```

```python
fig,ax = plt.subplots()
_=ax.set_xlim([-1.01,-0.99])
_=ax.set_ylim([0.99,1.01])
_=ax.plot(x,y)
```

```python
fig,ax = plt.subplots()
plt.axvline(x =0, color = 'lightgray')
plt.axhline(y=0,color = 'lightgray')
plt.xlim(-5,10)
plt.ylim(-10,80)
plt.plot(x,y)
plt.axvline(x=5,color = 'purple', linestyle = '--')
plt.axhline(y= 37,color = 'purple', linestyle ='--')
_ = ax.plot(x,y)
```

```python
def my_fxn(my_x):
  my_y = (my_x**2 - 1)/(my_x -1)
  return my_y
```

```python
#my_fxn(1)
```

```python
my_fxn(0.99)
```

```python
y = my_fxn(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0,color = 'lightgray')
plt.axhline(y=0,color = 'lightgray')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.axvline(x=1,color = 'purple', linestyle = '--')
plt.axhline(y=2,color = 'purple', linestyle = '--')
_=ax.plot(x,y)
```
