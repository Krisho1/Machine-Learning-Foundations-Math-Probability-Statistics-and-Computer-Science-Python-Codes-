dy/dx = f(x+delta_x) - f(x) / delta_x

```python
def my_deff(my_fxn,my_x,my_delta):
  return ((my_fxn(my_x + my_delta) - my_fxn(my_x))/(my_delta))
```

```python
deltas = [1,0.1,0.001,0.0001,0.00001,0.000000001,0.000000000000000000001]
```

```python
for delta in deltas:
  print(my_deff(my_fxn,2,delta))
```

```python
for delta in deltas:
  print(my_deff(my_fxn,-1,delta))
```

using tensor flow

```python
import tensorflow as tf
```

```python
x = tf.Variable(5.0)
x
```

```python
with tf.GradientTape() as t:
  t.watch(x)  #Forward pass
  y=x**2
```

```python
t.gradient(y,x)
```
